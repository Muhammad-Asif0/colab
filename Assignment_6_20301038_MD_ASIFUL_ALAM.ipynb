```
#task 2
class Dates:

    def __init__(self,date):
        self.date=date
    def getDate(self):
        return self.date
    @classmethod
    def toDashDate(cls,date):
        date=date.replace("/","-")
        obj=cls(date)
        return obj
date1 = Dates("05-09-2020")
dateFromDB = "05/09/2020"
date2= Dates.toDashDate(dateFromDB)
if(date1.getDate() == date2.getDate()):
    print("Equal")
else:
    print("Unequal")
```

```
##task 3

class Passenger:
    count=0
    def __init__(self,name):
        self.name=name
        self.weight=450
        Passenger.count+=1
    def set_bag_weight(self,weight):
        if weight<=20:
            self.weight=self.weight
        elif weight>=21 and weight<50:
            self.weight=self.weight+50
        else:
            self.weight=self.weight+100
    def printDetail(self):
        print("Name:",self.name)
        print("Bus Fare:",self.weight)
print("Total Passenger:", Passenger.count)
p1 = Passenger("Jack")
p1.set_bag_weight(90)
p2 = Passenger("Carol")
p2.set_bag_weight(10)
p3 = Passenger("Mike")
p3.set_bag_weight(25)
print("=========================")
p1.printDetail()
print("=========================")
p2.printDetail()
print("=========================")
p3.printDetail()
print("=========================")
print("Total Passenger:", Passenger.count)
```

```
#task 5

class Fruit:
    count=0
    fruitCount=0
    def __init__(self,name,number):
        self.name=name
        self.count=number
        Fruit.fruitCount+=number
    @classmethod
    def resetcount(cls,reset=0):
        cls.fruitCount=reset
    @staticmethod
    def saySomethingGood():
        print("Fruits are good for health")
apples = Fruit("Apple", 3)
pears = Fruit("Pear", 4)
print(apples.name, apples.count)
print(pears.name,pears.count)
print("Total number of fruits", Fruit.fruitCount)
Fruit.saySomethingGood()
Fruit.resetcount()
print("Total number of fruits", Fruit.fruitCount)
```

```
class Cat:
    Number_of_cats=0
    def __init__(self,colour,action):
        self.colour=colour
        self.action=action
        Cat.Number_of_cats+=1
    def printCat(self):
        print(self.colour,"cat is",self.action)
    @classmethod
    def no_parameter(cls):
        colour,action="White","sitting"
        obj=cls(colour,action)
        return obj
    @classmethod
    def first_parameter(cls,name):
        colour,action=name,"sitting"
        obj = cls(colour,action)
        return obj
    @classmethod
    def second_parameter(cls, action):
        action,colour = action,"Grey"
        obj = cls(colour,action)
        return obj
    def changeColor(self,colour):
        self.colour=colour
print("Total number of cats:", Cat.Number_of_cats)
c1 = Cat.no_parameter()
c2 = Cat.first_parameter("Black")
c3 = Cat("Brown", "jumping")
c4 = Cat("Red", "purring")
c5 = Cat.second_parameter("playing")
print("=======================")
c1.printCat()
c2.printCat()
c3.printCat()
c4.printCat()
c5.printCat()
c1.changeColor("Blue")
c3.changeColor("Purple")
c1.printCat()
c3.printCat()
print("=======================")
print("Total number of cats:", Cat.Number_of_cats)
```
