```
#task 1

class Calculator:
    def calculate(self,first_value=0,operator="",second_value=0):
        self.first_value=first_value
        self.operator=operator
        self.second_value=second_value
        print("Letâ€™s Calculate!")
        print("Value 1:",first_value)
        print("Operator:", operator)
        print("Value 2:", second_value)
        if operator=="+":
            print("Result: ",first_value+second_value)
        elif operator=="-":
            print("Result: ",first_value-second_value)
        elif operator=="*":
            print("Result: ",first_value*second_value)
        elif operator=="/":
            print("Result: ",first_value/second_value)
object=Calculator()
object.calculate(int(input()),input(),int(input()))
```

```
#task 2

class Customer:

    def __init__(self,name=None):
        self.name=name
    def greet(self,name=None):
        if name==None:
            print("Hello!")
        else:
            print("Hello",name)

    def purchase(self,*list):
        print(self.name+",", "you purchased", len(list) ,"item(s):")
        for i in list:
            print(i)
customer_1 = Customer("Sam")
customer_1.greet()
customer_1.purchase("chips", "chocolate", "orange juice")
print("-----------------------------")
customer_2 = Customer("David")
customer_2.greet("David")
customer_2.purchase("orange juice")
```

```
#task 3

class Panda:
    def __init__(self,name,gender,age):
        self.name=name
        self.gender=gender
        self.age=age
    def sleep(self,hours=""):
        if hours=="":
          return "{}'s duration is unknown thus should have only bamboo leaves".format(self.name)  
        elif int(hours)>=int(3) and int(hours)<=int(5):
          return "{} sleeps {} hours daily and should have Mixed Veggies".format(self.name,hours)
        elif int(hours)>=int(6) and int(hours)<=int(8):
          return "{} sleeps {} hours daily and should have Eggplant & Tofu".format(self.name,hours)
        elif int(hours)>=int(9) and int(hours)<=int(11):
          return "{} sleeps {} hours daily and should have Broccoli Chicken".format(self.name,hours)
panda1 = Panda("Kunfu", "Male", 5)
panda2 = Panda("Pan Pan", "Female", 3)
panda3 = Panda("Ming Ming", "Female", 8)
print("{} is a {} Panda Bear who is {} years old".format(panda1.name,panda1.gender,panda1.age))
print("{} is a {} Panda Bear who is {} years old".format(panda2.name,panda2.gender,panda2.age))
print("{} is a {} Panda Bear who is {} years old".format(panda3.name,panda3.gender,panda3.age))
print(panda2.sleep(10))
print(panda1.sleep(4))
print(panda3.sleep())

```

```
##task 5

class Vehicle:
    def __init__(self,x=0,y=0):
        self.x=x
        self.y=y
    def moveUp(self):
        self.y=self.y+1
    def moveLeft(self):
        self.x=self.x-1
    def moveDown(self):
        self.y=self.y-1
    def moveRight(self):
        self.x=self.x+1
    def print_position(self):
        print("(",self.x,",",self.y,")")
car = Vehicle()
car.print_position()
car.moveUp()
car.print_position()
car.moveLeft()
car.print_position()
car.moveDown()
car.print_position()
car.moveRight()
```

```
#task 7

class Student:
    def __init__(self,name,id,department="CSE"):
        self.name=name
        self.id=id
        self.department=department
    def dailyEffort(self,hour):
        print("NAME:", self.name)
        print("ID:",self.id)
        print("Department:",self.department)
        print("Daily Effort:",hour,"hour(s)")
        if hour <=2 :
            print('Suggestion: Should give more effort!')
        elif hour <=4:
            print('Suggestion: Keep up the good work!')
        else:
            print('Suggestion: Excellent! Now motivate others.')
    def printDetails(self,detail=""):
        if detail!="":
            print("Details:",detail)

harry = Student('Harry Potter', 123)
harry.dailyEffort(3)
harry.printDetails()
print('========================')
john = Student("John Wick", 456, "BBA")
john.dailyEffort(2)
john.printDetails()
print('========================')
naruto = Student("Naruto Uzumaki", 777, "Ninja")
naruto.dailyEffort(6)
naruto.printDetails()
```

```
##task 8

class Patient:
    def __init__(self,name,age,symptoms=""):
        self.name=name
        self.age=age
        self.symptoms=symptoms
    def add_Symptom(self,*symptom):
      if len(symptom)==1:
          self.symptoms=symptom[0]
      elif len(symptom)>1:
          self.symptoms=symptom[0]
          for i in range(1,len(symptom)):
              self.symptoms=self.symptoms+","+symptom[i]
    def printPatientDetail(self):
        print("Name:",self.name)
        print("Age:",self.age)
        print("symptom:", self.symptoms)
p1 = Patient('Thomas', 23)
p1.add_Symptom('Headache')
p2 = Patient('Carol', 20)
p2.add_Symptom('Vomiting', 'Coughing')
p3 = Patient('Mike', 25)
p3.add_Symptom('Fever', 'Headache', 'Coughing')
print("=========================")
p1.printPatientDetail()
print("=========================")
p2.printPatientDetail()
print("=========================")
p3.printPatientDetail()
print("=========================")
```

```
#task 9

class Avengers:
    def __init__(self,name,partner,Super_powers=""):
        self.name=name
        self.partner=partner
        self.Super_powers=Super_powers
    def super_powers(self,*powers):
        if len(powers) == 1:
            self.Super_powers= powers[0]
        elif len(powers) > 1:
            self.Super_powers =  powers[0]
            for i in range(1, len(powers)):
                self.Super_powers = self.Super_powers + "," + powers[i]

    def printAvengersDetail(self):
        print("Name:", self.name)
        print("Partner:", self.partner)
        print("Super powers:",self.Super_powers)
a1 = Avengers('Captain America', 'Bucky Barnes')
a1.super_powers('Stamina', 'Slowed ageing')
a2 = Avengers('Doctor Strange', 'Ancient One')
a2.super_powers('Mastery of magic')
a3 = Avengers('Iron Man', 'War Machine')
a3.super_powers('Genius level intellect', 'Scientist ')
print("=========================")
a1.printAvengersDetail()
print("=========================")
a2.printAvengersDetail()
print("=========================")
a3.printAvengersDetail()
print("=========================")
```
