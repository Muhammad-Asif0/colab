```
class Shape:
    def __init__(self,shape,one,two):
        self.one=one
        self.two=two
        self.shape=shape
    def area(self):
        if self.shape=="Square":
            print("Area:",self.one*self.two)
        elif self.shape=="Triangle":
            print("Area:",(self.one**2)+self.two)
        elif self.shape=="Rhombus":
            print("Area:",(self.one*self.two)/2)
        elif self.shape=="Rectangle":
            print("Area:",self.one*self.two)
        else:
            print("Shape unknown")
triangle_1 = Shape("Triangle",10,25)
triangle_1.area()
print("==========================")
square_1 = Shape("Square",10,10)
square_1.area()
print("==========================")
rhombus_1 = Shape("Rhombus",18,25)
rhombus_1.area()
print("==========================")
rectangle_1 = Shape("Rectangle",15,30)
rectangle_1.area()
print("==========================")
trapezium_1 = Shape("Trapezium",15,30)
trapezium_1.area()
```

•	The default <strong>fare charge</strong> of any vehicle is seating capacity * 100 and add an extra 10% on full fare as a maintenance charge. So total fare for a vehicle instance will become the final amount = total fare + 10% of the total fare.<br>

<br>
COPY THIS PORTION:<br>

<strong>OUTPUT:</strong><br>

Vehicle Name: School Volvo<br>
Speed: 180 <br>
Mileage: 12<br>
The seating capacity of a bus is 50 passengers<br>
Total Bus fare is: 5500.0<br>


```
class Vehicle:
    def __init__(self,name,speed,mileage):
        self.name=name
        self.speed=speed
        self.mileage=mileage
        self.passenger=0
        print("Vehicle Name:",self.name)
        print("Speed:",self.speed)
        print("Mileage:",self.mileage)
    def seating_capacity(self,default=50):
        self.passenger=default
        return "The seating capacity of a bus is "+str(default)+" passengers"
    def fare(self):
        print("Total Bus fare is:",self.passenger*100+((10*self.passenger*100)/100))


School_bus = Vehicle("School Volvo", 180, 12)
print(School_bus.seating_capacity())
print("Total Bus fare is:", School_bus.fare())
```

```
class Student:
    def __init__(self,name,id,dep,*list):
        self.name=name
        self.id=id
        self.cgpa=0.0
        self.dep=dep
        self.list=list
    def calculate_CGPA(self):
        for i in range (len(self.list[0])):
            self.cgpa+=self.list[0][i]
        self.cgpa=self.cgpa/len(self.list[0])
    def print_details(self):
        print("--------- Department:",self.dep,"---------")
        print("Name",self.name,",","ID",self.id)
        print("CGPA: "+"{:.2f}".format(self.cgpa))
        if self.cgpa>=3.80:
            print("Your academic standing is Highest Distinction")
        elif self.cgpa>=3.65 and self.cgpa<=3.79:
            print("Your academic standing is High Distinction")
        elif self.cgpa>=3.50 and self.cgpa<=3.64:
            print("Your academic standing is Distinction")
        elif self.cgpa>=2.00 and self.cgpa<=3.49:
            print("Your academic standing is Satisfactory")
        else:
            print("You cannot graduate")


s1 = Student('Dora', '15995599','CSE', [4,3.7,3.7,4])
s1.calculate_CGPA()
s1.print_details()

s2 = Student('Pingu', '12312322', 'EEE', [1.7,1.3,1.3,1.3,1])
s2.calculate_CGPA()
s2.print_details()

s3 = Student('Bob', '13311331', 'CSE', [2,3,3,3.7,2.7,2.7])
s3.calculate_CGPA()
s3.print_details()
```
